Implement Gradient Descent Algorithm to find the local minima of a function.
For example, find the local minima of the function y=(x+3)Â² starting from the point x=2.

```python
import numpy as np
```

```python
def GDA(initial_theta, learning_rate = 1e-4,max_iter = 100,stopping_threshold = 1e-6):
    
    def J(theta): return (theta+3)**2  #fn. to be optimized
    def gredient(theta): return (2*theta) + 6
    
    theta = initial_theta
    cost  = J(theta)
    
    values = {
                "weights": [theta],
                "costs": [cost]
             }
    
    print(f"Iteration 0:\t cost = {cost} | theta = {theta} ")

    for i in range(max_iter):
        
        theta = theta - ((learning_rate)*(gredient(theta)))
        curr_cost = J(theta)
        
        if cost and (cost - curr_cost) <= stopping_threshold:
            print("\nCost does not decrease further\nEarly stopping...\n")
            break
        
        cost = curr_cost
        
        values["weights"].append(theta)
        values["costs"].append(cost)
        
        print(f"Iteration {i+1}:\t cost = {cost} | theta = {theta}")
    return values
```

```python
values = GDA(initial_theta = 2,learning_rate = 1e-2,max_iter =  500)
# print(values)
```

```python
print(f"Local minimum of given function is  = {round(values['costs'][-1],6)} at x = {round(values['weights'][-1],6)}")
```
