```python
import pandas as pd
import numpy as np
import seaborn as sns
from statsmodels.graphics.mosaicplot import mosaic
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("datasets_600600_1109547_hospital_beds_per_india_v1.csv")
df.iloc[30:31,7:8] = (df.iloc[30:31,7:8]/1000)
df 
```

# Conclusion:
* The cell above shows the pandas library has sucessfully imported.
* Dataframe is created and the **.csv file** has read.The dataset shown is about country **India**.
* Dataframe has sucessfully displayed.

```python
df.dropna(axis = 1,inplace = True)
df.drop(['source','source_url','type','measure'],axis = 'columns',inplace = True)
df1 = np.split(df,[36],axis = 0)
india_states = df1[0]
india = df1[1]
india_states  # india_states as dataframe
```

```python
india #india as dataframe
```

# Conclusion:
* Above dataframes shows the cleaning of unnecessory data.The column with null values is discarded with the help of **pandas.dropna()** function.
* The last two columns(source and source_url) are discarded with **pandas.drop()** function.
* We have created two dataframes.One containing all states of india and other as india.

```python
st = india_states['state']
pop = india_states['population']

plt.figure(figsize=(7,9))

plt.xlabel("States")
plt.ylabel("Population")
plt.title("Population Per State")

plt.bar(st,pop,width=0.9,align='center')
plt.subplots_adjust(bottom=0.15)
plt.xticks(st,rotation =-90)

plt.show()
```

# Conclusion:
* The bar graph shows the population for each state during the year 2014-2018.

```python
byyear = india_states.groupby('year')
byear = byyear[['beds','population']].sum()
by_year = byear.reset_index()
by_year
```

# Conclusion:
* The dataframe is created in which data is grouped according to year.This is done using **df.groupby()** function.
* The above dataframe displayes the total number of beds per 1000 inhabitants and corresponding population per year.


```python
plt.figure(figsize=(9,3))

plt.subplot(121)
plt.xlabel("Year")
plt.ylabel("Beds Capacity")
plt.title("No.of Beds/year")
plt.plot(by_year.year,by_year.beds,linewidth =3)

plt.subplot(122)
plt.title("Population/year")
plt.xlabel("Year")
plt.ylabel("Population (10^8)")
plt.plot(by_year.year,by_year.population,'b')

plt.xticks(by_year["year"].tolist())
plt.subplots_adjust(wspace=0.3)
plt.grid()
plt.show()
```

# <div style="text-align:center">BEDS CAPACITY ANALYSIS FOR  COUNTRY USA </div>

```python
udf = pd.read_csv("datasets_600600_1109547_hospital_beds_USA_v1.csv",na_values={'beds':'0'})
udf
```

# Conclusion:
* The cell above shows the **pandas library** has sucessfully imported.
* Dataframe is created and the **.csv file** has read.The dataset shown is about country **USA**.
* The dataframe containing null values are replaced with **NAN**.Thus, missing data has successfully handled.
* Dataframe has successfully displayed.

```python
udf.drop(['measure','source','source_url'],axis = 1,inplace = True)
udf.rename(columns={'county':'city'} ,inplace = True)
udf.dropna(inplace=True)
udf.reset_index(drop = True,inplace = True)
```

```python
udf
```

# Conclusion:
* The above datafame shows the cleaned data.
* The files containing source and source url's has sucessfully removed.
* The NAN values are droped using **dropna()** function.also, the variable (county) name has updated.


```python
print('BED CAPACITY IN EVERY CITY OF USA (PER STATE)'.center(140,"*"))
bystate = udf.loc[:,('state','city','beds','year','population')].groupby('state')
state_list = list(udf['state'].unique())

for a,b, in bystate:
  state_list = list(udf['state'].unique())
  fig = plt.figure(figsize=(6,6))
  bycity = bystate.get_group(str(a)).groupby('city')
  bed_city = bycity.sum().head(10).reset_index()
  
  plt.xlabel("city")
  plt.ylabel("beds")
  plt.title(str(a))
  plt.bar(bed_city.city,bed_city.beds,width=0.9,align='center')
  plt.xticks(bed_city.city,rotation =90)
  plt.show()
```

```python
sns.lmplot(x ='population',y ='beds',data=udf,hue='year',palette="Set1",legend_out=False)
plt.title('Beds capacity (2012-2020) ')
plt.show()
```

# Conclusion:
* The above plot gives a visualization about the total beds in USA per its  population during year 2012-2020.
* We can see during 2018, with population around 2lac, their is a better situation for bed capacity.during 2019 the bed capacity has decreased.Thus, their is a need improvement.

```python
sp = bystate.mean().reset_index()
plt.figure(figsize=(15,5))
ax = sns.scatterplot(x="state", y='beds', hue="year",data=sp)
ax.set_title('Bed capacities per state for diffrent durations')
plt.xticks(sp['state'])
plt.show()
```

# Conclusion:
* The scatterplot shows the beds capacity in each state during it's source year.
* We can see the minimum capacity is in **The district of columbia** while, high capacity of beds is in **Kansas** state

```python
bytype = udf.loc[:,('type','beds')].groupby('type')
bp = bytype.mean().reset_index()
plt.figure(figsize = (10, 6))
ax = sns.barplot(x='type',y='beds',data=bp,ci='sd')
#plt.setp(ax.artists, alpha=.5, linewidth=2, edgecolor="k")
ax.set_title("Barplot for bed capacity in USA")
plt.yticks(fontsize = 12)
plt.xticks(rotation=45,fontsize = 12)
plt.show()
```

# Conclusion:
* The barplot shows the bed capacity for each type of beds.We observe that the **ACUTE** type has more responses as compare to other types of beds required.The **OTHER** type includes ‘SPECIAL’, ‘CHRONIC DISEASE’, ‘CHILDREN’, ‘LONG TERM CARE’, ‘REHABILITATION’, ‘WOMEN’, ‘MILITARY’ which has around 0.68 per 1000 inhabitants.

```python
plt.figure(figsize=(6,6))
sns.lineplot(x= 'year',y = 'beds',hue='type',data=udf,palette="Set1", lw=2,style='type',estimator=None)
plt.title("Lineplot for types of beds in USA")
plt.legend(loc='upper left')
plt.grid(alpha=0.5)
plt.show()
```

# Conclusion:
* We can observe that, during year 2018-19, the beds capacity is highest.while, by the time of 2020, it has suddenly decreased due due to COVID-19.thus, their is a need to increase the production rate by the manufacturing companies.

# <div style="text-align:center"><u>COMPARATIVE BED ANALYSIS BETWEEN INDIA AND USA</u></div>


```python
stt_india = india_states[['beds','population']].describe()
stt_india_bed = stt_india.iloc[:,:1].reset_index()
ind = pd.Series(['india']*8,name="country")
bed_conty = pd.concat([stt_india_bed,ind],axis=1)

bystate = udf.loc[:,('state','beds','population')].groupby('state') 
stt_usa = bystate.sum().describe()
stt_usa_bed =stt_usa.iloc[:,:1].reset_index()
ind = pd.Series(['USA']*8,name="country")
bed_conty_1 = pd.concat([stt_usa_bed,ind],axis=1)

final_comp = pd.concat([bed_conty,bed_conty_1])
plt.title("Beds capacity between india and USA ")
sns.lineplot(x='index',y='beds',hue='country',data=final_comp)
plt.show()
```

# Conclusion:
 * Above lineplot shows comparitive analysis of beds capacity between countries India and USA during year 2012-2020.
 Although, their is a count difference in states of two countries, We come to the conclusion that in 2018-19, USA has more bed capacity than india.The maxima and minima for both the countries are differing a lot.
 Thus, their is a need to fill this gap for the betterment of future.
